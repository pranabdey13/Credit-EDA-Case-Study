<h4>BINS

<h6>Income_Range

```python
def Income(x):
    
    if x < 100000:
        return "0-100000"
    elif 100000 <= x < 200000:
        return "100000-200000"
    elif 200000 <= x < 300000:
        return "200000-300000"
    elif 300000 <= x < 400000:
        return "300000-400000"
    else:
        return ">400000"


inp0['Income_Range'] = inp0.AMT_INCOME_TOTAL.apply(lambda x: Income(x))
```

```python
inp0.Income_Range.head()
```

Creating an AGE column by calculating age DAYS_BIRTH

```python
inp0['DAYS_BIRTH']=abs(inp0['DAYS_BIRTH'])
```

```python
inp0['DAYS_BIRTH_YEAR']= abs(inp0['DAYS_BIRTH']/365).astype(int)
```

```python
def Age(x):
    
    if x < 18:
        return "child"
    elif 18 <= x < 25:
        return "Teenager"
    elif 25 <= x < 45:
        return "Young"
    elif 45 <= x < 100:
        return "Senior"
    else:
        return "unknown"


inp0['Age_Slot'] = inp0.DAYS_BIRTH_YEAR.apply(lambda x: Age(x))
```
